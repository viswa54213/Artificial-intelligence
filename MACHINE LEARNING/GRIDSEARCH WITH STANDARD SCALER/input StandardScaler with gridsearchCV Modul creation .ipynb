```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset, dtype=int, drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[[ 'charges']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#doing standardscaler for inputs caz of minimum and maximum value have large difference value
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
X_test
```

```python
#creating module with gridsearch applying the standardscalered inputs 
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid= {'kernel':['rbf','linear','poly','sigmoid'],'gamma':['scale', 'auto'],'C':[10, 50, 100]}
grid=GridSearchCV(SVR(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(X_train,Y_train)

```

```python
print("best parameter combo{}:".format(grid.best_params_))
```

```python
re=grid.cv_results_
table=pd.DataFrame.from_dict(re)
table
```

```python
prediction=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(prediction,Y_test)
r_score
```

```python
import pickle 
filename="final.sav"
```

```python
pickle.dump(grid,open(filename,"wb"))
```
