```python
#pandas is a one of the library ,its use to read the database like excel and number related files , 
#it is used for separating the input and output data set
import pandas as pd
```

```python
#here weare reading the dataset from the inbuild function of pandas
dataset=pd.read_csv("Salary_Data.csv") 
```

```python
# readed dataset
dataset 
```

```python
#then here we are separating the input and output with the variables
independent=dataset[["YearsExperience"]]
dependent=dataset[["Salary"]]
independent
dependent
```

```python
#sklearn is also one of the library ,we are going to use the libray in every machine learnin project ,its like a default 
#library for ML
#here we are separating the input and output dataset for training and testing prpose ,we are taking 70% for training and 30% for testing testing
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
X_train,X_test,Y_train,Y_test
```

```python
#the main think is here , we are creating a model here 
#by using the sklear library of function . LinearRegression algorithm . y=wX+b<>module created  with input and output training dataset
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
regressor
```

```python
#then here we are checking the value of w and b ,by using sklearn library
weight=regressor.coef_
weight
baias=regressor.intercept_
baias
```

```python
#here we are predicting the value of teste input with the created module
y_predict=regressor.predict(X_test)
```

```python
#here we are checking errors inbtween teste input value predicted by our modul and test output
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_predict)
r_score
```

```python
#here we are saving the best module with the picklelibrary
import pickle
filename="final_model_linear.sav"
pickle.dump(regressor,open(filename,'wb'))
```

```python
# here we are checking our module how it's working again once
loaded_model=pickle.load(open("final_model_linear.sav",'rb'))
```
