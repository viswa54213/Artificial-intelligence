```python
import pandas as pd
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
# taking the best inputs rows and columns for finding the clusters
X=dataset.iloc[:,[3,4]].values
```

```python
X
```

```python
import matplotlib.pyplot as plt
```

```python
# finding the clusters in given dataset
from sklearn.cluster import KMeans
list=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i ,init='k-means++', random_state=42)
    kmeans.fit(X)
    list.append(kmeans.inertia_)
plt.plot(range(1,11),list)
plt.title('The elow method')
plt.xlabel('clusters')
plt.ylabel('wcss') 
plt.show()
```

```python
#creating modulefor cluster
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmean=kmeans.fit_predict(X)
```

```python
y_kmean
```

```python
#adding the clusterd valuse in columns
supervised=pd.DataFrame(dataset)
```

```python
supervised['clusterd group']=y_kmean
```

```python
supervised
```

```python
#saving the files as csv
supervised.to_csv('clusters.csv',index=False)
```
