```python
import pandas as pd
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
# taking the best inputs rows and columns for finding the clusters
X=dataset.iloc[:,[3,4]].values
```

```python
X
```

```python
import matplotlib.pyplot as plt
```

```python

import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendrogram')
plt.xlabel('customers')
plt.ylabel('Euclidean distance') 
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=5)
label=cluster.fit_predict(X)
```

```python
label
```

```python
#adding the clusterd valuse in columns
supervised=pd.DataFrame(dataset)
```

```python
supervised['clusterd group']=label
```

```python
supervised
```

```python
#saving the files as csv
supervised.to_csv(' Hierarchical AgglomerativeClustering.csv',index=False)
```
