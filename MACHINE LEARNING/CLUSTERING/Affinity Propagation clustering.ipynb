```python
import pandas as pd
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
# taking the best inputs rows and columns for finding the clusters
X=dataset.iloc[:,[3,4]].values
```

```python
X
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import AffinityPropagation
affin=AffinityPropagation(random_state=0)
label=affin.fit_predict(X)
```

```python
label
```

```python
#adding the clusterd valuse in columns
supervised=pd.DataFrame(dataset)
```

```python
supervised[' cluster group']=label
```

```python
supervised
```

```python
#saving the files as csv
supervised.to_csv('Affinity Propagation clustering.csv',index=False)
```
