```python
#pandas is a one of the library ,its use to read the database like excel and number related files , 
#it is used for separating the input and output data set
import pandas as pd
dataset=pd.read_csv("insurance_pre.csv")
```

```python
# readed dataset
dataset
```

```python
# herewe are changing the catagrical input to numerical input 
# new function
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
# this is a new function for reading the input and output names
#then here we are separating the input and output with the variables
dataset.columns
```

```python
#then here we are separating the input and output with the variables
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[["charges"]]
independent
dependent
```

```python
#creating module for RandomForestRegressor applying the parameters with the hepl of grid search
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
param_grid = {'n_estimators':[10, 20],'criterion':['squared_error', 'absolute_error'],'max_features':['sqrt', 'log2']}
grid = GridSearchCV(RandomForestRegressor(), param_grid, refit=True, verbose=3, n_jobs=-1)
grid.fit(independent, dependent)
```

```python
#seeing best combination parameters
print("best parameters{}:".format(grid.best_params_))
```

```python
#viewing the all combinations parameters modules in table
re=grid.cv_results_
table=pd.DataFrame.from_dict(re)
table
```

```python
future_prediction=grid.predict([[26,33.70,3,1,0]])
print("future prediction{}:".format(future_prediction))
```
