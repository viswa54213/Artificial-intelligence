```python
#pandas is a one of the library ,its use to read the database like excel and number related files , 
#it is used for separating the input and output data set
import pandas as pd
```

```python
#here weare reading the dataset from the inbuild function of pandas
dataset=pd.read_csv("insurance_pre.csv")
```

```python
# readed dataset
dataset
```

```python
# herewe are changing the catagrical input to numerical input 
# new function
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
# this is a new function for reading the input and output names
dataset.columns
```

```python
#then here we are separating the input and output with the variables
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[["charges"]]
independent
dependent
```

```python
#sklearn is also one of the library ,we are going to use the libray in every machine learnin project ,its like a default 
#library for ML
#here we are separating the input and output dataset for training and testing prpose ,we are taking 70% for training and 30% for testing testing
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
X_train,X_test,Y_train,Y_test
```

```python
#the main think is here , we are creating a model here WITH applying parameters with GRIDSEARCHCV
#by using the sklear library of function . LinearRegression algorithm . y=wX+b<>module created  with input and output training dataset
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
param_grid={'fit_intercept':[True,False],'n_jobs':[10,20,30,100],'positive':[True,False]}
grid=GridSearchCV(LinearRegression(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(X_train,Y_train)
grid
```

```python
#seeing best combination parameters
re=grid.cv_results_
print("besrt parameters{}:".format(grid.best_params_))
```

```python
#viewing the all combinations parameters modules in table
table=pd.DataFrame.from_dict(re)
table
```

```python
#here we are predicting the value of teste input with the created module
grid_prediction=grid.predict(X_test)
```

```python
#here we are checking errors inbtween teste input value predicted by our modul and test output
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,grid_prediction)
r_score
```
